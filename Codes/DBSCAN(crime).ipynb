```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('D:/study docs/Excel R/Data science/assignment/Clustering/Datasets/crime_data.csv')
```

```python
data.head()
```

```python
data = data.drop(['Unnamed: 0'],axis = 1)
```

```python
data_array = data.values
data_array
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_array)
```

```python
data_scaled
```

```python
dbscan = DBSCAN(eps = 0.8,min_samples = 4)
dbscan.fit(data_scaled)
```

```python
dbscan.labels_
```

```python
data['clusters'] = dbscan.labels_
```

```python
data
```

```python
sns.scatterplot(x = 'clusters',y = 'Assault',data = data)
```

```python
sns.barplot(x = 'clusters', y = 'Murder',data =data)
```
