```python
# import hierarchical clustering libraries
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sn
```

```python
data = pd.read_csv('D:/study docs/Excel R/Data science/assignment/Clustering/Datasets/crime_data.csv')
```

```python
data = data.drop(['Unnamed: 0'],axis = 1)
```

```python
data_array = data.values
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data_array)
```

```python
dendrogram = sch.dendrogram(sch.linkage(scaled_data, method='complete'))
```

```python
hc = AgglomerativeClustering(n_clusters=2,affinity ='euclidean',linkage = 'complete')
```

```python
y_hc = hc.fit_predict(scaled_data)
```

```python
data['Cluster'] =hc.labels_
```
